## Equation solver

```python
@jit
def SchrodingerEq_solver(hamiltonian, psi0, time_duration, dt, hbar=1.0):
    """
    Solves the time-dependent Schrdinger equation (TDSE).
    
    Args:
        hamiltonian_fn: Callable, H(t), a function that returns the Hamiltonian at time t.
        psi0: jnp.array, initial state vector.
        t_span: tuple, (t_start, t_end) the time range.
        dt: float, time step.
        hbar: float, Planck's constant divided by 2 (default: 1.0).
        
    Returns:
        t_vals: jnp.array, array of time points.
        psi_t: jnp.array, array of state vectors at each time point.
    """
    def time_derivative(psi, t):
        H_t = hamiltonian(t)  # Evaluate the Hamiltonian at time t
        return -1j / hbar * jnp.dot(H_t, psi)
    
    # Time points
    t_start, t_end = t_span
    t_vals = jnp.arange(t_start, t_end, dt)
    
    # Initialize state vector storage
    psi_t = [psi0]
    
    # Time evolution using Runge-Kutta 4th order method
    for t in t_vals[:-1]:
        k1 = time_derivative(psi_t[-1], t)
        k2 = time_derivative(psi_t[-1] + 0.5 * dt * k1, t + 0.5 * dt)
        k3 = time_derivative(psi_t[-1] + 0.5 * dt * k2, t + 0.5 * dt)
        k4 = time_derivative(psi_t[-1] + dt * k3, t + dt)
        psi_next = psi_t[-1] + (dt / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)
        psi_t.append(psi_next)
    
    return t_vals, jnp.array(psi_t)

@partial(jax.jit, static_argnums=(0, 1))
def lindblad_solver(hamiltonian, jump_operators, rho0, t_span, dt, hbar=1.0):
    """
    Solves the Lindblad master equation.
    
    Args:
        hamiltonian_fn: Callable, H(t), a function that returns the Hamiltonian at time t.
        jump_operators: list of jnp.ndarray, jump operators {L_k}.
        rho0: jnp.ndarray, initial density matrix.
        t_span: tuple, (t_start, t_end), the time range.
        dt: float, time step.
        hbar: float, Planck's constant divided by 2 (default: 1.0).
    
    Returns:
        t_vals: jnp.ndarray, array of time points.
        rho_t: jnp.ndarray, array of density matrices at each time point.
    """
    def lindblad_rhs(rho, t):
        """
        Calculates the right-hand side of the Lindblad equation.
        """
        H = hamiltonian_fn(t)
        
        # Commutator term: -i/hbar * [H, rho]
        commutator = -1j / hbar * (jnp.dot(H, rho) - jnp.dot(rho, H))
        
        # Lindblad dissipation terms
        dissipation = jnp.zeros_like(rho, dtype=jnp.complex64)
        for L in jump_operators:
            L_rho_Ldag = jnp.dot(L, jnp.dot(rho, L.conj().T))
            Ldag_L = jnp.dot(L.conj().T, L)
            dissipation += L_rho_Ldag - 0.5 * (jnp.dot(Ldag_L, rho) + jnp.dot(rho, Ldag_L))
        
        return commutator + dissipation
    
    # Time evolution using RK4
    def rk4_step(rho, t):
        k1 = lindblad_rhs(rho, t)
        k2 = lindblad_rhs(rho + 0.5 * dt * k1, t + 0.5 * dt)
        k3 = lindblad_rhs(rho + 0.5 * dt * k2, t + 0.5 * dt)
        k4 = lindblad_rhs(rho + dt * k3, t + dt)
        return rho + (dt / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)
    
    # Time points
    t_start, t_end = t_span
    t_vals = jnp.arange(t_start, t_end, dt)
    
    # Initialize storage for density matrices
    rho_t = [rho0]
    
    # Time evolution loop
    for t in t_vals[:-1]:
        rho_next = rk4_step(rho_t[-1], t)
        rho_t.append(rho_next)
    
    return t_vals, jnp.array(rho_t)
```

```python
import jax
import jax.numpy as jnp
from functools import partial

def tdse_solver(hamiltonian_fn, psi0, t_span, dt, hbar=1.0):
    """
    Solves the time-dependent Schrdinger equation (TDSE) using Runge-Kutta 4th order.
    
    Args:
        hamiltonian_fn: Callable, H(t), a function that returns the Hamiltonian at time t.
        psi0: jnp.array, initial state vector.
        t_span: tuple, (t_start, t_end), the time range.
        dt: float, time step.
        hbar: float, Planck's constant divided by 2 (default: 1.0).
    
    Returns:
        t_vals: jnp.array, array of time points.
        psi_t: jnp.array, array of state vectors at each time point.
    """
    def time_derivative(psi, t):
        H_t = hamiltonian_fn(t)  # Evaluate the Hamiltonian at time t
        return -1j / hbar * jnp.dot(H_t, psi)

    def rk4_step(psi, t):
        k1 = time_derivative(psi, t)
        k2 = time_derivative(psi + 0.5 * dt * k1, t + 0.5 * dt)
        k3 = time_derivative(psi + 0.5 * dt * k2, t + 0.5 * dt)
        k4 = time_derivative(psi + dt * k3, t + dt)
        return psi + (dt / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)
    
    t_start, t_end = t_span
    t_vals = jnp.arange(t_start, t_end, dt)
    psi_t = [psi0]
    
    for t in t_vals[:-1]:
        psi_next = rk4_step(psi_t[-1], t)
        psi_t.append(psi_next)
    
    return t_vals, jnp.array(psi_t)

# : 
def qubit_hamiltonian(t, omega_q=1.0, omega_d=1.0, A=0.1, phi=0.0):
    """
    Returns the time-dependent Hamiltonian for a single qubit under microwave drive.
    """
    sigma_x = jnp.array([[0, 1], [1, 0]], dtype=jnp.complex64)
    sigma_z = jnp.array([[1, 0], [0, -1]], dtype=jnp.complex64)
    
    H0 = 0.5 * omega_q * sigma_z  # 
    H_drive = A * jnp.cos(omega_d * t + phi) * sigma_x  # 
    return H0 + H_drive

#  |0
psi0 = jnp.array([1.0, 0.0], dtype=jnp.complex64)

# 
t_span = (0.0, 10.0)
dt = 0.01

# 
t_vals, psi_t = tdse_solver(lambda t: qubit_hamiltonian(t), psi0, t_span, dt)

# 
import matplotlib.pyplot as plt

plt.plot(t_vals, jnp.abs(psi_t[:, 0])**2, label=r"$|0\rangle$")
plt.plot(t_vals, jnp.abs(psi_t[:, 1])**2, label=r"$|1\rangle$")
plt.xlabel("Time")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
import jax.numpy as jnp
from jax import jit, vmap
import plotly.graph_objects as go
import numpy as np

# Wigner
def displacement_operator(alpha, dim):
    """
    
    alpha: 
    dim: 
    """
    n = jnp.arange(dim)
    coeff = jnp.exp(-0.5 * jnp.abs(alpha)**2) * (alpha**n) / jnp.sqrt(factorial(n))
    D = jnp.outer(coeff, coeff.conj())
    return D

@jit
def wigner_function(rho, x, y, dim):
    """
    Wigner
    rho:  (dim x dim)
    x, y: 
    dim: 
    """
    alpha = x + 1j * y
    D_alpha = displacement_operator(-alpha, dim)
    parity_operator = jnp.diag([1 if i % 2 == 0 else -1 for i in range(dim)])
    W = jnp.trace(rho @ D_alpha @ parity_operator @ D_alpha.conj().T).real
    return W

# :  |0><0|
def vacuum_state(dim):
    state = jnp.zeros((dim, 1))
    state = state.at[0, 0].set(1.0)  # |0> 
    rho = state @ state.conj().T
    return rho

# Wigner
def compute_wigner(rho, x_range, y_range, dim):
    x = jnp.linspace(*x_range, 100)
    y = jnp.linspace(*y_range, 100)
    X, Y = jnp.meshgrid(x, y)
    
    # 
    W = vmap(lambda x, y: wigner_function(rho, x, y, dim), in_axes=(0, 0))
    W_values = W(X.ravel(), Y.ravel())
    return X, Y, W_values.reshape(X.shape)

# Plotly
def plot_wigner(X, Y, W):
    fig = go.Figure(data=go.Contour(
        x=X[0, :],
        y=Y[:, 0],
        z=W,
        colorscale="Viridis",
        contours=dict(showlines=False),
    ))
    fig.update_layout(
        title="Wigner Function",
        xaxis_title="Re(\u03b1)",
        yaxis_title="Im(\u03b1)",
        xaxis=dict(scaleanchor="y"),
        yaxis=dict(scaleanchor="x"),
    )
    fig.show()

# 
def main():
    dim = 10  # 
    rho = vacuum_state(dim)  # 
    x_range = (-3, 3)
    y_range = (-3, 3)

    X, Y, W = compute_wigner(rho, x_range, y_range, dim)
    plot_wigner(X, Y, W)

if __name__ == "__main__":
    main()
```
