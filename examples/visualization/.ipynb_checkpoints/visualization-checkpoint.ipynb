```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from scipy.stats import multivariate_normal
```

```python
def vec2bloch(state_vec: jax.Array) -> jax.Array:
    '''

    '''
    # extract coefficients \alpha, \beta
    alpha = state_vec[0].item()
    beta = state_vec[1].item()
    #
    x = 2 * jnp.real(alpha*jnp.conj(beta))
    y = 2 * jnp.imag(alpha*jnp.conj(beta))
    z = jnp.abs(alpha)**2 - jnp.abs(beta)**2

    return jnp.array([x, y, z])
```

```python
def plot(state_vec: jax.Array, fig_size: tuple) -> None:
    '''
    Plots a single-qubit state vector on the Bloch sphere.

    Args:
      state_vec (jax.Array):
        The single-qubit state vector to plot. It must be a complex,
        array-like object (e.g., NumPy or JAX array) with shape (2,).
        The vector is expected to be normalized (i.e., its norm is 1).

    Returns:
      tuple[matplotlib.figure.Figure, matplotlib.axes.Axes]:
        A tuple containing the newly created Figure and 3D Axes objects,
        allowing for further customization after the function call.

    Raises:
      ValueError: If the input "state_vec" is not a 2-element vector.
      ValueError: If the input "state_vec" is not normalized.
    '''
    # matplotlib settings
    fig = plt.figure(figsize=fig_size)
    ax = fig.add_subplot(1, 1, 1, projection="3d")

    # plot settings
    ## convert cartesian coordinates to　polar coordinate
    phi = np.linspace(0, np.pi, 1000)
    theta = np.linspace(0, 2*np.pi, 1000)
    phi, theta = np.meshgrid(phi, theta)

    ## axis settings
    x_axis = np.sin(phi) * np.cos(theta)
    y_axis = np.sin(phi) * np.sin(theta)
    z_axis = np.cos(phi)

    ## plot bloch sphere
    #ax.plot_surface(x_axis, y_axis, z_axis, color='skyblue', alpha=0.1, linewidth=0, antialiased=True)
    ax.set_aspect("equal")

    ## axis range settings
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])

    # arrow settings
    arrow_length = 1.0 
    arrow_ratio = 0.1 
    ## x-axis arrow
    ax.quiver(0, 0, 0, arrow_length, 0, 0, color="tab:orange", arrow_length_ratio=arrow_ratio)
    ax.quiver(0, 0, 0, -arrow_length, 0, 0, color="tab:orange", arrow_length_ratio=arrow_ratio)
    ax.text(arrow_length*1.05, 0, 0, r"$| + \rangle$", color='red', fontsize=12)
    ax.text(-arrow_length*1.2, 0, 0, r"$| - \rangle$", color='red', fontsize=12)

    ## y-axis arrow
    ax.quiver(0, 0, 0, 0, arrow_length, 0, color="tab:green", arrow_length_ratio=arrow_ratio)
    ax.quiver(0, 0, 0, 0, -arrow_length, 0, color="tab:green", arrow_length_ratio=arrow_ratio)
    ax.text(0, arrow_length*1.2, 0, r"$| -i \rangle$", color='green', fontsize=12)
    ax.text(0, -arrow_length*1.2, 0, r"$| +i \rangle$", color='green', fontsize=12)

    ## z-axis arrow
    ax.quiver(0, 0, 0, 0, 0, arrow_length, color="tab:blue", arrow_length_ratio=arrow_ratio)
    ax.quiver(0, 0, 0, 0, 0, -arrow_length, color="tab:blue", arrow_length_ratio=arrow_ratio)
    ax.text(-0.075, 0, arrow_length*1.2, r"$| 0 \rangle$", color='blue', fontsize=12)
    ax.text(-0.075, 0, -arrow_length*1.2, r"$| 1 \rangle$", color='blue', fontsize=12)
    
    # convert state vector to bloch vector
    bloch_vec = vec2bloch(state_vec)
    ## convert jax.array to numpy.ndarray
    bloch_vec = jax.device_get(bloch_vec)

    ax.quiver(0, 0, 0, 
              bloch_vec[0], bloch_vec[1], bloch_vec[2],
              color="black", 
              arrow_length_ratio=0.1,
              zorder=2)

    ax.set_title("Single-qubit state on Bloch sphere")
    ax.grid(False)
    ax.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
def plot_2d(density_mat: jax.Array, fig_size: tuple) -> None:
    '''

    '''
    # matplotlib settings
    fig = plt.figure(figsize=fig_size)
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    ax1.set_title("Real part")
    ax2.set_title("Imaginary part")

    # aspect settings
    ax1.set_aspect('equal')
    ax2.set_aspect('equal')

    # extract real&imaginary part in density matrix
    real_part = jnp.real(density_mat)
    imag_part = jnp.imag(density_mat)

    # plot real&imaginary part
    hm1 = ax1.pcolormesh(real_part, cmap="viridis")
    hm2 = ax2.pcolormesh(imag_part, cmap="cividis")

    # colorbar settings
    fig.colorbar(hm1, ax=ax1)
    fig.colorbar(hm2, ax=ax2)

    plt.tight_layout(rect=[0, 0.03, 0.75, 0.95])
    #plt.tight_layout()
    plt.show()
```

```python
n = 10
x = np.linspace(-2, 2, n)
y = np.linspace(-2, 2, n)
X, Y = np.meshgrid(x, y)

# 複素行列の例
Z = np.exp(-(X**2 + Y**2)/2) * np.exp(1j * (X + Y))
plot_2d(Z, (13, 4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d(density_mat: jax.Array, fig_size: tuple) -> None:
    '''

    '''
    # matplotlib settings
    fig = plt.figure(figsize=fig_size)
    ax1 = fig.add_subplot(1, 2, 1, projection="3d")
    ax2 = fig.add_subplot(1, 2, 2, projection="3d")
    ax1.set_title("Real part")
    ax2.set_title("Imaginary part")

    # extract matrix size
    row_size = density_mat.shape[0]
    col_size = density_mat.shape[1]
    
    # extract real&imaginary part in density matrix
    real_part = jnp.real(density_mat)
    imag_part = jnp.imag(density_mat)

    # plot real&imaginary part
    for i in range(row_size):
        for j in range(col_size):
            ax1.bar3d(i, j, 0, 0.8, 0.8, real_part[i, j], color="tab:blue" if real_part[i, j] >= 0 else "tab:orange")
            ax2.bar3d(i, j, 0, 0.8, 0.8, imag_part[i, j], color="tab:purple" if imag_part[i, j] >= 0 else "tab:green")

    value_min = min(np.min(real_part), np.min(imag_part))
    value_max = max(np.max(real_part), np.max(imag_part))
    ax1.set_zlim(value_min, value_max)
    ax2.set_zlim(value_min, value_max)

    plt.tight_layout(rect=[0, 0.03, 0.75, 0.95])
    #plt.tight_layout()
    plt.show()
```

```python
n = 10
x = np.linspace(-2, 2, n)
y = np.linspace(-2, 2, n)
X, Y = np.meshgrid(x, y)

# 複素行列の例
Z = np.exp(-(X**2 + Y**2)/2) * np.exp(1j * (X + Y))
plot_3d(Z, (13, 4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy.special import hermite, factorial
import scipy.integrate as integrate

# Planck constant (for simplicity, let ħ = 1)
hbar = 1.0

def psi_harmonic_oscillator(x, n, omega=1.0):
    """
    Harmonic oscillator eigenfunctions (using Hermite polynomials)
    n: quantum number (0, 1, 2, ...)
    omega: oscillator frequency
    """
    # Harmonic oscillator parameter
    alpha = np.sqrt(omega / hbar)
    
    # Normalization constant
    normalization = (alpha / np.sqrt(np.pi * 2**n * factorial(n)))**0.5
    
    # Hermite polynomial
    hermite_poly = hermite(n)
    
    return normalization * np.exp(-alpha * x**2 / 2) * hermite_poly(alpha * x)

def wigner_harmonic_oscillator(x, p, n, omega=1.0):
    """
    Analytical solution for Wigner function of harmonic oscillator
    n: quantum number (0, 1, 2, ...)
    """
    # Dimensionless phase space coordinates
    alpha = np.sqrt(omega / hbar)
    X = alpha * x
    P = p / (hbar * alpha)
    
    # Squared radius in phase space
    r_squared = X**2 + P**2
    
    # Laguerre polynomial calculation
    L_n = np.polynomial.laguerre.Laguerre.basis(n)
    laguerre_eval = L_n(2 * r_squared)
    
    # Wigner function
    W = (-1)**n / (np.pi * hbar) * np.exp(-r_squared) * laguerre_eval
    
    return W

def calculate_wigner_grid(wigner_function, x_range, p_range):
    """
    Calculate Wigner function values on a grid
    
    Parameters:
    -----------
    wigner_function : function
        Function that takes (x, p) and returns W(x, p)
    x_range : array
        Array of position values
    p_range : array
        Array of momentum values
        
    Returns:
    --------
    X, P : meshgrid
        Meshgrid of position and momentum values
    W : array
        2D array of Wigner function values
    """
    X, P = np.meshgrid(x_range, p_range)
    W = np.zeros_like(X)
    
    for i in range(len(x_range)):
        for j in range(len(p_range)):
            try:
                # Calculate Wigner function value and catch any numerical errors
                val = wigner_function(x_range[i], p_range[j])
                # Replace any NaN or Inf values with 0
                if np.isnan(val) or np.isinf(val):
                    W[j, i] = 0.0
                else:
                    W[j, i] = val
            except:
                # In case of any exception, use 0
                W[j, i] = 0.0
    
    # Convert from jnp array to numpy array if needed
    if not isinstance(W, np.ndarray):
        W = np.array(W)
    
    # Final check for NaN or Inf values
    W[np.isnan(W)] = 0.0
    W[np.isinf(W)] = 0.0
        
    return X, P, W

def plot_wigner_3d(wigner_function, 
                   x_range=np.linspace(-4, 4, 100), 
                   p_range=np.linspace(-4, 4, 100), 
                   title="Wigner Function",
                   colormap=cm.coolwarm, 
                   elev=30, azim=30, 
                   figsize=(10, 8)):
    """
    Create a 3D surface plot of a Wigner function
    
    Parameters:
    -----------
    wigner_function : function
        Function that takes (x, p) and returns W(x, p)
    x_range : array
        Array of position values
    p_range : array
        Array of momentum values
    title : str
        Plot title
    colormap : matplotlib colormap
        Colormap for the surface
    elev, azim : float
        Elevation and azimuth angles for 3D view
    figsize : tuple
        Figure size
        
    Returns:
    --------
    fig, ax : matplotlib figure and axis
        The figure and axis objects for further customization
    """
    # Calculate Wigner function on grid
    X, P, W = calculate_wigner_grid(wigner_function, x_range, p_range)
    
    # Check for empty or invalid data
    if np.all(W == 0):
        print("Warning: All Wigner function values are zero or invalid.")
        
    # Create 3D plot
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot surface
    surf = ax.plot_surface(X, P, W, cmap=colormap, linewidth=0, antialiased=True)
    
    # Set labels and title
    ax.set_xlabel('Position x')
    ax.set_ylabel('Momentum p')
    ax.set_zlabel('W(x,p)')
    ax.set_title(title)
    
    # Set view angle
    ax.view_init(elev=elev, azim=azim)
    
    # Set z-axis symmetrically to emphasize negative values
    # Use a small default if all values are near zero
    absmax = np.max(np.abs(W))
    if absmax < 1e-10:
        absmax = 1.0
    ax.set_zlim(-absmax, absmax)
    
    # Add colorbar
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
    
    plt.tight_layout()
    
    return fig, ax

def plot_wigner_2d(wigner_function, 
                   x_range=np.linspace(-4, 4, 100), 
                   p_range=np.linspace(-4, 4, 100), 
                   title="Wigner Function Heatmap",
                   colormap=plt.cm.RdBu_r, 
                   figsize=(10, 8)):
    """
    Create a 2D heatmap of a Wigner function
    
    Parameters:
    -----------
    wigner_function : function
        Function that takes (x, p) and returns W(x, p)
    x_range : array
        Array of position values
    p_range : array
        Array of momentum values
    title : str
        Plot title
    colormap : matplotlib colormap
        Colormap for the heatmap (default: RdBu_r for red-white-blue)
    figsize : tuple
        Figure size
        
    Returns:
    --------
    fig, ax : matplotlib figure and axis
        The figure and axis objects for further customization
    """
    # Calculate Wigner function on grid
    X, P, W = calculate_wigner_grid(wigner_function, x_range, p_range)
    
    # Check for empty or invalid data
    if np.all(W == 0):
        print("Warning: All Wigner function values are zero or invalid.")
    
    # Create 2D plot
    fig, ax = plt.subplots(figsize=figsize)
    
    # Create heatmap with symmetric color scale (centered at 0)
    vmax = np.max(np.abs(W))
    # Use a small default if all values are near zero
    if vmax < 1e-10:
        vmax = 1.0
    im = ax.pcolormesh(X, P, W, cmap=colormap, shading='auto', vmin=-vmax, vmax=vmax)
    
    # Set labels and title
    ax.set_xlabel('Position x')
    ax.set_ylabel('Momentum p')
    ax.set_title(title)
    
    # Add colorbar
    cbar = fig.colorbar(im, ax=ax)
    cbar.set_label('W(x,p)')
    
    plt.tight_layout()
    
    return fig, ax

# Example usage
if __name__ == "__main__":
    # Define range for position and momentum
    x_range = np.linspace(-4, 4, 100)
    p_range = np.linspace(-4, 4, 100)
    
    # Ground state
    def wigner_ground_state(x, p):
        return wigner_harmonic_oscillator(x, p, n=0)
    
    # First excited state
    def wigner_first_excited(x, p):
        return wigner_harmonic_oscillator(x, p, n=1)
    
    # Superposition state (|0⟩ + |1⟩)/√2
    def wigner_superposition(x, p):
        w0 = wigner_harmonic_oscillator(x, p, n=0)
        w1 = wigner_harmonic_oscillator(x, p, n=1)
        # Avoid potential numerical issues with the square root
        product = w0 * w1
        if product < 0:
            sqrt_term = 0  # Avoid complex numbers
        else:
            sqrt_term = np.sqrt(product)
        return 0.5 * (w0 + w1) + np.real(np.exp(1j * np.pi/4) * sqrt_term)
    
    # Example 1: 3D plot of ground state
    plot_wigner_3d(wigner_ground_state, 
                  title="Ground State Wigner Function (n=0)", 
                  colormap=cm.viridis)
    
    # Example 2: 2D plot of first excited state
    plot_wigner_2d(wigner_first_excited,
                  title="First Excited State Wigner Function (n=1)")
    
    # Example 3: Both plots for superposition state
    plot_wigner_3d(wigner_superposition,
                  title="Superposition State Wigner Function (|0⟩ + |1⟩)/√2",
                  colormap=cm.coolwarm)
    
    plot_wigner_2d(wigner_superposition,
                  title="Superposition State Wigner Function - Negative Values (Red) Show Quantum Interference")
    
    plt.show()
```
