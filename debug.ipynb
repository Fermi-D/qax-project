```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy.special import hermite, factorial
import scipy.integrate as integrate

# Planck constant (for simplicity, let ħ = 1)
hbar = 1.0

def psi_harmonic_oscillator(x, n, omega=1.0):
    """
    Harmonic oscillator eigenfunctions (using Hermite polynomials)
    n: quantum number (0, 1, 2, ...)
    omega: oscillator frequency
    """
    # Harmonic oscillator parameter
    alpha = np.sqrt(omega / hbar)
    
    # Normalization constant
    normalization = (alpha / np.sqrt(np.pi * 2**n * factorial(n)))**0.5
    
    # Hermite polynomial
    hermite_poly = hermite(n)
    
    return normalization * np.exp(-alpha * x**2 / 2) * hermite_poly(alpha * x)

def wigner_harmonic_oscillator(x, p, n, omega=1.0):
    """
    Analytical solution for Wigner function of harmonic oscillator
    n: quantum number (0, 1, 2, ...)
    """
    # Dimensionless phase space coordinates
    alpha = np.sqrt(omega / hbar)
    X = alpha * x
    P = p / (hbar * alpha)
    
    # Squared radius in phase space
    r_squared = X**2 + P**2
    
    # Laguerre polynomial calculation
    L_n = np.polynomial.laguerre.Laguerre.basis(n)
    laguerre_eval = L_n(2 * r_squared)
    
    # Wigner function
    W = (-1)**n / (np.pi * hbar) * np.exp(-r_squared) * laguerre_eval
    
    return W

def wigner_superposition(x, p, psi_func, y_range=(-10, 10), num_points=1000):
    """
    Calculate Wigner function for arbitrary wavefunction using numerical integration
    psi_func: wavefunction (function returning complex values)
    """
    def integrand(y):
        return psi_func(x + y/2).conjugate() * psi_func(x - y/2) * np.exp(1j * p * y / hbar)
    
    # Range for numerical integration
    y = np.linspace(y_range[0], y_range[1], num_points)
    dy = y[1] - y[0]
    
    # Evaluate the integrand
    integrands = np.array([integrand(y_val) for y_val in y])
    
    # Numerical integration (trapezoidal rule)
    integral = np.trapz(integrands, dx=dy)
    
    # Normalization constant
    wigner = integral / (np.pi * hbar)
    
    return wigner.real  # Wigner function should be real-valued

# Grid setup
x_range = np.linspace(-4, 4, 100)
p_range = np.linspace(-4, 4, 100)
X, P = np.meshgrid(x_range, p_range)

# Calculate Wigner functions for different quantum states
W_ground = np.zeros_like(X)
W_first = np.zeros_like(X)
W_superposition = np.zeros_like(X)

# Calculate Wigner function at each point on the grid
for i in range(len(x_range)):
    for j in range(len(p_range)):
        # Ground state (n=0)
        W_ground[j, i] = wigner_harmonic_oscillator(x_range[i], p_range[j], n=0)
        
        # First excited state (n=1)
        W_first[j, i] = wigner_harmonic_oscillator(x_range[i], p_range[j], n=1)
        
        # Superposition state: (|0⟩ + |1⟩)/√2

# Superposition of ground and first excited states
W_superposition = 0.5 * (W_ground + W_first) + \
                  np.real(np.exp(1j * np.pi/4) * np.sqrt(W_ground * W_first))

# 3D plot setup
fig = plt.figure(figsize=(18, 6))
plt.suptitle('Wigner Functions of Harmonic Oscillator', fontsize=16)

# Wigner function for ground state (n=0)
ax1 = fig.add_subplot(131, projection='3d')
surf1 = ax1.plot_surface(X, P, W_ground, cmap=cm.viridis, 
                         linewidth=0, antialiased=True)
ax1.set_title('Ground State (n=0)')
ax1.set_xlabel('Position x')
ax1.set_ylabel('Momentum p')
ax1.set_zlabel('W(x,p)')

# Wigner function for first excited state (n=1)
ax2 = fig.add_subplot(132, projection='3d')
surf2 = ax2.plot_surface(X, P, W_first, cmap=cm.plasma, 
                         linewidth=0, antialiased=True)
ax2.set_title('First Excited State (n=1)')
ax2.set_xlabel('Position x')
ax2.set_ylabel('Momentum p')
ax2.set_zlabel('W(x,p)')

# Wigner function for superposition state
ax3 = fig.add_subplot(133, projection='3d')
surf3 = ax3.plot_surface(X, P, W_superposition, cmap=cm.coolwarm, 
                         linewidth=0, antialiased=True)
ax3.set_title('Superposition State (|0⟩ + |1⟩)/√2')
ax3.set_xlabel('Position x')
ax3.set_ylabel('Momentum p')
ax3.set_zlabel('W(x,p)')

# Adjust viewpoint
for ax in [ax1, ax2, ax3]:
    ax.view_init(elev=30, azim=30)
    # Set Z-axis range symmetrically to emphasize negative values
    zmin = min(W_ground.min(), W_first.min(), W_superposition.min())
    zmax = max(W_ground.max(), W_first.max(), W_superposition.max())
    absmax = max(abs(zmin), abs(zmax))
    ax.set_zlim(-absmax, absmax)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# 2D heatmap visualization of Wigner functions (focus on negative values)
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
plt.suptitle('Wigner Function Heatmaps - Negative Values (Red) Indicate Quantum Interference', fontsize=16)

# Color map setting (white at center, blue for positive, red for negative)
cmap = plt.cm.RdBu_r

# Heatmap for ground state
im1 = axs[0].pcolormesh(X, P, W_ground, cmap=cmap, shading='auto', 
                        vmin=-np.max(np.abs(W_ground)), vmax=np.max(np.abs(W_ground)))
axs[0].set_title('Ground State (n=0)')
axs[0].set_xlabel('Position x')
axs[0].set_ylabel('Momentum p')
fig.colorbar(im1, ax=axs[0])

# Heatmap for first excited state
im2 = axs[1].pcolormesh(X, P, W_first, cmap=cmap, shading='auto',
                        vmin=-np.max(np.abs(W_first)), vmax=np.max(np.abs(W_first)))
axs[1].set_title('First Excited State (n=1)')
axs[1].set_xlabel('Position x')
axs[1].set_ylabel('Momentum p')
fig.colorbar(im2, ax=axs[1])

# Heatmap for superposition state
im3 = axs[2].pcolormesh(X, P, W_superposition, cmap=cmap, shading='auto',
                        vmin=-np.max(np.abs(W_superposition)), vmax=np.max(np.abs(W_superposition)))
axs[2].set_title('Superposition State (|0⟩ + |1⟩)/√2')
axs[2].set_xlabel('Position x')
axs[2].set_ylabel('Momentum p')
fig.colorbar(im3, ax=axs[2])

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy.special import hermite, factorial
import scipy.integrate as integrate

# Planck constant (for simplicity, let ħ = 1)
hbar = 1.0

def psi_harmonic_oscillator(x, n, omega=1.0):
    """
    Harmonic oscillator eigenfunctions (using Hermite polynomials)
    n: quantum number (0, 1, 2, ...)
    omega: oscillator frequency
    """
    # Harmonic oscillator parameter
    alpha = np.sqrt(omega / hbar)
    
    # Normalization constant
    normalization = (alpha / np.sqrt(np.pi * 2**n * factorial(n)))**0.5
    
    # Hermite polynomial
    hermite_poly = hermite(n)
    
    return normalization * np.exp(-alpha * x**2 / 2) * hermite_poly(alpha * x)

def wigner_harmonic_oscillator(x, p, n, omega=1.0):
    """
    Analytical solution for Wigner function of harmonic oscillator
    n: quantum number (0, 1, 2, ...)
    """
    # Dimensionless phase space coordinates
    alpha = np.sqrt(omega / hbar)
    X = alpha * x
    P = p / (hbar * alpha)
    
    # Squared radius in phase space
    r_squared = X**2 + P**2
    
    # Laguerre polynomial calculation
    L_n = np.polynomial.laguerre.Laguerre.basis(n)
    laguerre_eval = L_n(2 * r_squared)
    
    # Wigner function
    W = (-1)**n / (np.pi * hbar) * np.exp(-r_squared) * laguerre_eval
    
    return W

def calculate_wigner_grid(wigner_function, x_range, p_range):
    """
    Calculate Wigner function values on a grid
    
    Parameters:
    -----------
    wigner_function : function
        Function that takes (x, p) and returns W(x, p)
    x_range : array
        Array of position values
    p_range : array
        Array of momentum values
        
    Returns:
    --------
    X, P : meshgrid
        Meshgrid of position and momentum values
    W : array
        2D array of Wigner function values
    """
    X, P = np.meshgrid(x_range, p_range)
    W = np.zeros_like(X)
    
    for i in range(len(x_range)):
        for j in range(len(p_range)):
            try:
                # Calculate Wigner function value and catch any numerical errors
                val = wigner_function(x_range[i], p_range[j])
                # Replace any NaN or Inf values with 0
                if np.isnan(val) or np.isinf(val):
                    W[j, i] = 0.0
                else:
                    W[j, i] = val
            except:
                # In case of any exception, use 0
                W[j, i] = 0.0
    
    # Convert from jnp array to numpy array if needed
    if not isinstance(W, np.ndarray):
        W = np.array(W)
    
    # Final check for NaN or Inf values
    W[np.isnan(W)] = 0.0
    W[np.isinf(W)] = 0.0
        
    return X, P, W

def plot_wigner_3d(wigner_function, 
                   x_range=np.linspace(-4, 4, 100), 
                   p_range=np.linspace(-4, 4, 100), 
                   title="Wigner Function",
                   colormap=cm.coolwarm, 
                   elev=30, azim=30, 
                   figsize=(10, 8)):
    """
    Create a 3D surface plot of a Wigner function
    
    Parameters:
    -----------
    wigner_function : function
        Function that takes (x, p) and returns W(x, p)
    x_range : array
        Array of position values
    p_range : array
        Array of momentum values
    title : str
        Plot title
    colormap : matplotlib colormap
        Colormap for the surface
    elev, azim : float
        Elevation and azimuth angles for 3D view
    figsize : tuple
        Figure size
        
    Returns:
    --------
    fig, ax : matplotlib figure and axis
        The figure and axis objects for further customization
    """
    # Calculate Wigner function on grid
    X, P, W = calculate_wigner_grid(wigner_function, x_range, p_range)
    
    # Check for empty or invalid data
    if np.all(W == 0):
        print("Warning: All Wigner function values are zero or invalid.")
        
    # Create 3D plot
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot surface
    surf = ax.plot_surface(X, P, W, cmap=colormap, linewidth=0, antialiased=True)
    
    # Set labels and title
    ax.set_xlabel('Position x')
    ax.set_ylabel('Momentum p')
    ax.set_zlabel('W(x,p)')
    ax.set_title(title)
    
    # Set view angle
    ax.view_init(elev=elev, azim=azim)
    
    # Set z-axis symmetrically to emphasize negative values
    # Use a small default if all values are near zero
    absmax = np.max(np.abs(W))
    if absmax < 1e-10:
        absmax = 1.0
    ax.set_zlim(-absmax, absmax)
    
    # Add colorbar
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
    
    plt.tight_layout()
    
    return fig, ax

def plot_wigner_2d(wigner_function, 
                   x_range=np.linspace(-4, 4, 100), 
                   p_range=np.linspace(-4, 4, 100), 
                   title="Wigner Function Heatmap",
                   colormap=plt.cm.RdBu_r, 
                   figsize=(10, 8)):
    """
    Create a 2D heatmap of a Wigner function
    
    Parameters:
    -----------
    wigner_function : function
        Function that takes (x, p) and returns W(x, p)
    x_range : array
        Array of position values
    p_range : array
        Array of momentum values
    title : str
        Plot title
    colormap : matplotlib colormap
        Colormap for the heatmap (default: RdBu_r for red-white-blue)
    figsize : tuple
        Figure size
        
    Returns:
    --------
    fig, ax : matplotlib figure and axis
        The figure and axis objects for further customization
    """
    # Calculate Wigner function on grid
    X, P, W = calculate_wigner_grid(wigner_function, x_range, p_range)
    
    # Check for empty or invalid data
    if np.all(W == 0):
        print("Warning: All Wigner function values are zero or invalid.")
    
    # Create 2D plot
    fig, ax = plt.subplots(figsize=figsize)
    
    # Create heatmap with symmetric color scale (centered at 0)
    vmax = np.max(np.abs(W))
    # Use a small default if all values are near zero
    if vmax < 1e-10:
        vmax = 1.0
    im = ax.pcolormesh(X, P, W, cmap=colormap, shading='auto', vmin=-vmax, vmax=vmax)
    
    # Set labels and title
    ax.set_xlabel('Position x')
    ax.set_ylabel('Momentum p')
    ax.set_title(title)
    
    # Add colorbar
    cbar = fig.colorbar(im, ax=ax)
    cbar.set_label('W(x,p)')
    
    plt.tight_layout()
    
    return fig, ax

# Example usage
if __name__ == "__main__":
    # Define range for position and momentum
    x_range = np.linspace(-4, 4, 100)
    p_range = np.linspace(-4, 4, 100)
    
    # Ground state
    def wigner_ground_state(x, p):
        return wigner_harmonic_oscillator(x, p, n=0)
    
    # First excited state
    def wigner_first_excited(x, p):
        return wigner_harmonic_oscillator(x, p, n=1)
    
    # Superposition state (|0⟩ + |1⟩)/√2
    def wigner_superposition(x, p):
        w0 = wigner_harmonic_oscillator(x, p, n=0)
        w1 = wigner_harmonic_oscillator(x, p, n=1)
        # Avoid potential numerical issues with the square root
        product = w0 * w1
        if product < 0:
            sqrt_term = 0  # Avoid complex numbers
        else:
            sqrt_term = np.sqrt(product)
        return 0.5 * (w0 + w1) + np.real(np.exp(1j * np.pi/4) * sqrt_term)
    
    # Example 1: 3D plot of ground state
    plot_wigner_3d(wigner_ground_state, 
                  title="Ground State Wigner Function (n=0)", 
                  colormap=cm.viridis)
    
    # Example 2: 2D plot of first excited state
    plot_wigner_2d(wigner_first_excited,
                  title="First Excited State Wigner Function (n=1)")
    
    # Example 3: Both plots for superposition state
    plot_wigner_3d(wigner_superposition,
                  title="Superposition State Wigner Function (|0⟩ + |1⟩)/√2",
                  colormap=cm.coolwarm)
    
    plot_wigner_2d(wigner_superposition,
                  title="Superposition State Wigner Function - Negative Values (Red) Show Quantum Interference")
    
    plt.show()
```
